    }
}

### MCQs

**1) Which pair is a valid overload?**
- A) `int f(int a)` and `int f(int x)`
- B) `int f(int a)` and `double f(int a)`
- **C) `int f(int a)` and `double f(double a)` ✓**
- D) `int f(int a, int b)` and `int f(int a, int b)`

**Answer: C** - Only option C has different parameter types (int vs double), which creates a valid overload.

**2) Which is the AP CSA notion of a method signature?**
- A) `public static int max(int a, int b)`
- **B) `max(int, int)` ✓**
- C) `max(int a, int b) throws Exception`
- D) `max(int, double)`

**Answer: B** - A method signature in AP CSA consists of the method name and parameter types only (no modifiers, return type, or parameter names).

**3) With the overloads `void p(int x)` and `void p(double x)`, what prints?**
```java
p(5);
p(5.0);
```
- **A) `int`, `double` ✓**
- B) `double`, `double`
- C) `int`, `int`
- D) Compile error

**Answer: A** - `p(5)` calls the int version since 5 is an int literal. `p(5.0)` calls the double version since 5.0 is a double literal.

**4) Which call is ambiguous with only these two methods?**
```java
void h(long x) {}
void h(float x) {}
```
- **A) `h(5)` ✓**
- B) `h(5.0)`
- C) `h(5f)`
- D) None are ambiguous

**Answer: A** - `h(5)` is ambiguous because int can be widened to both long and float, and neither conversion is more specific than the other.

**5) Which statement is TRUE?**
- A) Parameter names affect the method signature.
- B) Return type affects the method signature.
- C) Access modifiers affect the method signature.
- **D) Overloading requires different parameter lists. ✓**

**Answer: D** - Overloading requires different parameter types, order, or arity (count). Parameter names, return types, and access modifiers do NOT affect the signature.


### Short answer

**Question 1: Explain why `int sum(int a, int b)` and `double sum(int a, int b)` cannot both exist.**

**Answer:** These two methods cannot both exist because they have the exact same signature `sum(int, int)`. In Java, the method signature includes only the method name and parameter types/order, NOT the return type. Since both methods have identical signatures, the compiler cannot distinguish between them, resulting in a compilation error. To overload methods, you must change the parameter types, order, or count—not just the return type.

**Question 2: In one sentence, distinguish parameters vs arguments.**

**Answer:** Parameters are the variables declared in a method's definition that specify what types of values the method accepts, while arguments are the actual values passed to the method when it is called.

**FRQ 1: Overloaded `indexOf` methods**

```java
public class IndexOfFRQ {
    /**
     * Returns the index of the first occurrence of target character in string s.
     * Returns -1 if not found.
     * 
     * Assumptions:
     * - s is not null
     * - Case-sensitive search
     */
    public static int indexOf(char target, String s) {
        for (int i = 0; i < s.length(); i++) {
            if (s.charAt(i) == target) {
                return i;
            }
        }
        return -1;
    }
    
    /**
     * Returns the index of the first occurrence of target substring in string s.
     * Returns -1 if not found.
     * 
     * Assumptions:
     * - Neither s nor target are null
     * - If target is empty, returns 0
     * - Case-sensitive search
     * 
     * Constraints:
     * - Cannot use String.indexOf() method
     * - Uses manual substring comparison
     */
    public static int indexOf(String target, String s) {
        if (target.length() == 0) return 0;
        if (target.length() > s.length()) return -1;
        
        for (int i = 0; i <= s.length() - target.length(); i++) {
            boolean match = true;
            for (int j = 0; j < target.length(); j++) {
                if (s.charAt(i + j) != target.charAt(j)) {
                    match = false;
                    break;
                }
            }
            if (match) {
                return i;
            }
        }
        return -1;
    }
    
    public static void main(String[] args) {
        // Test char version
        System.out.println(indexOf('l', "hello"));        // 2
        System.out.println(indexOf('x', "hello"));        // -1
        System.out.println(indexOf('h', "hello"));        // 0
        
        // Test String version
        System.out.println(indexOf("lo", "hello"));       // 3
        System.out.println("---");
        System.out.println(indexOf("xyz", "hello"));      // -1
        System.out.println(indexOf("", "hello"));         // 0
        System.out.println(indexOf("hello", "hello"));    // 0
        System.out.println(indexOf("Java", "I love Java"));  // 7
    }
}
```

**FRQ 2: Overloaded `clamp` methods**

```java
public class ClampFRQ {
    /**
     * Returns value constrained to the range [low, high].
     * If value < low, returns low.
     * If value > high, returns high.
     * Otherwise returns value.
     * 
     * Handles case where low > high by swapping them.
     */
    public static int clamp(int value, int low, int high) {
        // Swap if low > high
        if (low > high) {
            int temp = low;
            low = high;
            high = temp;
        }
        
        if (value < low) return low;
        if (value > high) return high;
        return value;
    }
    
    /**
     * Returns value constrained to the range [low, high] for doubles.
     * If value < low, returns low.
     * If value > high, returns high.
     * Otherwise returns value.
     * 
     * Handles case where low > high by swapping them.
     */
    public static double clamp(double value, double low, double high) {
        // Swap if low > high
        if (low > high) {
            double temp = low;
            low = high;
            high = temp;
        }
        
        if (value < low) return low;
        if (value > high) return high;
        return value;
    }
    
    public static void main(String[] args) {
        // Test int version
        System.out.println(clamp(15, 0, 10));      // 10 (clamped to max)
        System.out.println(clamp(-5, 0, 10));      // 0  (clamped to min)
        System.out.println(clamp(5, 0, 10));       // 5  (within range)
        System.out.println(clamp(50, 100, 10));    // 50 (low/high swapped, 50 within [10,100])
        
        System.out.println("---");
        
        // Test double version
        System.out.println(clamp(15.5, 0.0, 10.0));    // 10.0
        System.out.println(clamp(-2.7, 0.0, 10.0));    // 0.0
        System.out.println(clamp(7.3, 0.0, 10.0));     // 7.3
        System.out.println(clamp(5.5, 10.0, 0.0));     // 5.5 (swapped: [0,10])
    }
}
```

### Running FRQ 1: IndexOf Overloads


````python
public class IndexOfFRQ {
    public static int indexOf(char target, String s) {
        for (int i = 0; i < s.length(); i++) {
            if (s.charAt(i) == target) {
                return i;
            }
        }
        return -1;
    }
    
    public static int indexOf(String target, String s) {
        if (target.length() == 0) return 0;
        if (target.length() > s.length()) return -1;
        
        for (int i = 0; i <= s.length() - target.length(); i++) {
            boolean match = true;
            for (int j = 0; j < target.length(); j++) {
                if (s.charAt(i + j) != target.charAt(j)) {
                    match = false;
                    break;
                }
            }
            if (match) {
                return i;
            }
        }
        return -1;
    }
    
    public static void main(String[] args) {
        System.out.println("Testing char version:");
        System.out.println("indexOf('l', \"hello\") = " + indexOf('l', "hello"));
        System.out.println("indexOf('x', \"hello\") = " + indexOf('x', "hello"));
        System.out.println("indexOf('h', \"hello\") = " + indexOf('h', "hello"));
        
        System.out.println("\nTesting String version:");
        System.out.println("indexOf(\"lo\", \"hello\") = " + indexOf("lo", "hello"));
        System.out.println("indexOf(\"xyz\", \"hello\") = " + indexOf("xyz", "hello"));
        System.out.println("indexOf(\"\", \"hello\") = " + indexOf("", "hello"));
        System.out.println("indexOf(\"hello\", \"hello\") = " + indexOf("hello", "hello"));
        System.out.println("indexOf(\"Java\", \"I love Java\") = " + indexOf("Java", "I love Java"));
    }
}

IndexOfFRQ.main(null);
````

### Summary of Key Learnings

1. **Method Signature** = Method name + parameter types/order (NOT return type, parameter names, or modifiers)
2. **Overloading** requires different parameter lists (type, order, or count)
3. **Cannot overload** by changing only return type
4. **Type matching** at call site: Java selects the most specific applicable method
